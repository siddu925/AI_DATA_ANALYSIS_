```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Let's create a DataFrame with some potential outliers
data_outliers = {'Value': [20, 22, 25, 23, 21, 75, 24, 26, 28, 10, 90, 27, 29, 31, 5]}
df_outliers = pd.DataFrame(data_outliers)

print("Original DataFrame with potential outliers:\n", df_outliers)

# 1. Create a boxplot using seaborn
sns.boxplot(x=df_outliers['Value'])
plt.title('Boxplot to Detect Outliers')
plt.xlabel('Value')
plt.show()

# Let's create a DataFrame with outliers in multiple columns
data_multi_outliers = {'Feature A': [1, 2, 3, 4, 5, 100],
                       'Feature B': [6, 7, 8, 9, 10, -5],
                       'Feature C': [11, 12, 13, 14, 15, 50]}
df_multi_outliers = pd.DataFrame(data_multi_outliers)
print("\nDataFrame with potential outliers in multiple columns:\n", df_multi_outliers)

# Create boxplots for multiple columns
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_multi_outliers)
plt.title('Boxplots for Multiple Features')
plt.ylabel('Value')
plt.show()
```
