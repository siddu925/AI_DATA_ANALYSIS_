```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2 / 10
y4 = np.log(x + 1)

# Method 1: Using plt.subplot()
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, first subplot
plt.plot(x, y1, color='blue', linestyle='-')
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(x)')

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, second subplot
plt.plot(x, y2, color='red', linestyle='--')
plt.title('Cosine Wave')
plt.xlabel('X')
plt.ylabel('cos(x)')

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, third subplot
plt.plot(x, y3, color='green', linestyle=':')
plt.title('Quadratic Function')
plt.xlabel('X')
plt.ylabel('x^2 / 10')

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, fourth subplot
plt.plot(x, y4, color='purple', linestyle='-.')
plt.title('Logarithmic Function')
plt.xlabel('X')
plt.ylabel('log(x+1)')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.suptitle('Combined Subplots', fontsize=16, fontweight='bold', y=1.02) # Add a main title
plt.show()

# Method 2: Using plt.subplots() (more flexible)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Combined Subplots (using subplots)', fontsize=16, fontweight='bold')

# Plot on the first subplot (top-left)
axes[0, 0].plot(x, y1, color='blue', linestyle='-')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('sin(x)')

# Plot on the second subplot (top-right)
axes[0, 1].plot(x, y2, color='red', linestyle='--')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('cos(x)')

# Plot on the third subplot (bottom-left)
axes[1, 0].plot(x, y3, color='green', linestyle=':')
axes[1, 0].set_title('Quadratic Function')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('x^2 / 10')

# Plot on the fourth subplot (bottom-right)
axes[1, 1].plot(x, y4, color='purple', linestyle='-.')
axes[1, 1].set_title('Logarithmic Function')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('log(x+1)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for suptitle
plt.show()

# Method 3: Creating subplots with different sizes using GridSpec
import matplotlib.gridspec as gridspec

plt.figure(figsize=(14, 10))
gs = gridspec.GridSpec(3, 2, width_ratios=[1, 2], height_ratios=[1, 1, 2])
plt.suptitle('Combined Subplots with GridSpec', fontsize=16, fontweight='bold')

ax1 = plt.subplot(gs[0, 0])
ax1.plot(x, y1, color='blue')
ax1.set_title('Sine Wave')

ax2 = plt.subplot(gs[0, 1])
ax2.plot(x, y2, color='red')
ax2.set_title('Cosine Wave')

ax3 = plt.subplot(gs[1, 0])
ax3.plot(x, y3, color='green')
ax3.set_title('Quadratic')

ax4 = plt.subplot(gs[1, 1])
ax4.plot(x, y4, color='purple')
ax4.set_title('Logarithmic')

ax5 = plt.subplot(gs[2, :]) # Span across both columns in the third row
ax5.plot(x, y1, label='Sine', color='blue')
ax5.plot(x, y2, label='Cosine', color='red')
ax5.legend()
ax5.set_title('Combined Sine and Cosine')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
