```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Visualize the distribution of petal length using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of sepal width using a density plot (KDE)
plt.figure(figsize=(8, 6))
sns.kdeplot(iris['sepal_width'], fill=True)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()

# Visualize the distribution of petal width using a rug plot combined with KDE
plt.figure(figsize=(8, 6))
sns.kdeplot(iris['petal_width'], fill=True)
sns.rugplot(iris['petal_width'], color=".2")
plt.title('Distribution of Petal Width with Rug Plot')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Density')
plt.show()

# Visualize the distribution of sepal length using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(y=iris['sepal_length'])
plt.title('Box Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
plt.show()

# Visualize the distribution of sepal length by species using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()

# Visualize the distribution of petal length by species using a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
