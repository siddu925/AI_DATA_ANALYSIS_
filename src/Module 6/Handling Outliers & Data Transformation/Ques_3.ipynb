```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Let's create a DataFrame with a skewed distribution
data_skewed = {'Value': [1, 2, 3, 4, 5, 10, 15, 20, 30, 50]}
df_skewed = pd.DataFrame(data_skewed)

print("Original Skewed DataFrame:\n", df_skewed)

# Visualize the original distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_skewed['Value'], kde=True)
plt.title('Original Distribution')

plt.subplot(1, 2, 2)
stats.probplot(df_skewed['Value'], plot=plt)
plt.title('Probability Plot (Original)')
plt.tight_layout()
plt.show()

# 1. Apply Box-Cox Transformation
# The boxcox function returns the transformed data and the optimal lambda value
transformed_value, optimal_lambda = stats.boxcox(df_skewed['Value'])

# Create a new DataFrame with the transformed values and lambda
df_transformed = pd.DataFrame({'Transformed_Value': transformed_value})
print("\nDataFrame after Box-Cox Transformation:\n", df_transformed)
print("\nOptimal Lambda () for Box-Cox:", optimal_lambda)

# Visualize the transformed distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_transformed['Transformed_Value'], kde=True)
plt.title('Distribution after Box-Cox')

plt.subplot(1, 2, 2)
stats.probplot(df_transformed['Transformed_Value'], plot=plt)
plt.title('Probability Plot (Box-Cox)')
plt.tight_layout()
plt.show()

# Applying Box-Cox to a DataFrame column
df_skewed['Value_BoxCox'], optimal_lambda_col = stats.boxcox(df_skewed['Value'])
print("\nOriginal DataFrame with Box-Cox Transformed Value:\n", df_skewed)
print("\nOptimal Lambda for DataFrame Column:", optimal_lambda_col)
```
