```python
import pandas as pd
from sklearn.datasets import load_iris

# Part 1: Load a Dataset & Check Missing Values
# Task 1: Iris Dataset
iris_data = load_iris(as_frame=True)
iris_df = iris_data.frame
missing_values_iris = iris_df.isnull().sum().sum()
print(f"Total number of missing values in the Iris dataset: {missing_values_iris}\n")

# Part 2: Identify Duplicates & Inconsistencies
# Task 2: Inconsistent Entries in a Sample Dataset
# Let's create a sample DataFrame with a 'Gender' column for demonstration
sample_data = {'ID': [1, 2, 3, 4, 5],
               'Gender': ['Male', 'F', 'male', 'FEMALE', 'M']}
sample_df = pd.DataFrame(sample_data)

inconsistent_genders = sample_df['Gender'].unique()
print("Unique entries in the 'Gender' column (sample data):", inconsistent_genders)

# To identify truly inconsistent entries, you might want to standardize them.
# For example, converting all to lowercase and then checking for a standard set.
standardized_genders = sample_df['Gender'].str.lower().unique()
print("Unique entries in 'Gender' after converting to lowercase:", standardized_genders, "\n")

# Part 3: Generate a Data Quality Report
# Task 3: Iris Dataset Summary
print("--- Data Quality Report: Iris Dataset ---")
print(f"Total number of rows: {len(iris_df)}")
print(f"Total number of columns: {len(iris_df.columns)}")
print(f"\nMissing Values per Column:")
print(iris_df.isnull().sum())
print("\nBasic Descriptive Statistics:")
print(iris_df.describe())
print("--- End of Report ---")
```
