```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the figure and axes
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sine Wave',
         color='forestgreen',       # Custom color name
         linewidth=2,             # Line thickness
         linestyle='--',          # Dashed line style
         marker='o',              # Circle markers at data points
         markersize=5,            # Size of the markers
         markerfacecolor='lightcoral', # Fill color of the markers
         markeredgecolor='black',    # Outline color of the markers
         alpha=0.7)                # Transparency of the line

plt.plot(x, y2, label='Cosine Wave',
         color='#FF8C00',       # Hex color code (dark orange)
         linewidth=1.5,
         linestyle='-',           # Solid line style
         marker='^',              # Triangle markers
         markersize=6,
         markerfacecolor='skyblue',
         markeredgecolor='navy',
         alpha=0.9)

# Add labels and title
plt.xlabel('X-axis', fontsize=12, color='dimgray')
plt.ylabel('Y-axis', fontsize=12, color='dimgray')
plt.title('Sine and Cosine Waves', fontsize=14, fontweight='bold')

# Add a legend
plt.legend(fontsize=10, loc='upper right', frameon=False)

# Customize the grid
plt.grid(True, linestyle=':', alpha=0.5)

# Add annotations
plt.annotate('Local Maximum (Sine)', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10)

# Set axis limits
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)

# Customize ticks
plt.xticks(np.arange(0, 11, 1), fontsize=9)
plt.yticks(np.arange(-1, 1.1, 0.5), fontsize=9)

# Remove spines (axis borders)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add a background color
plt.gca().set_facecolor('#F0F0F0')

# Show the plot
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
