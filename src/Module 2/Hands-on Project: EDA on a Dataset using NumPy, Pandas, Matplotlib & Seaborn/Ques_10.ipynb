```python
# 1. Install the required libraries.
# You can run these commands in your terminal or Jupyter Notebook cell:
# pip install pandas numpy matplotlib seaborn

# 2. Verify the installation by importing the libraries.
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
print("Matplotlib version:", matplotlib.__version__)
print("Seaborn version:", sns.__version__)

print("\nEnvironment setup complete. Libraries imported successfully.")

# Step 2: Load & Explore the Dataset
# Objective: Load data into a pandas DataFrame and obtain a basic understanding of its structure.

# Create a dummy CSV file for demonstration
data = {'col1': [1, 2, 3, 4, 5],
        'col2': [5.5, 6.6, 7.7, 8.8, np.nan],
        'col3': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)
df.to_csv('sample_data.csv', index=False)

# 3. Load a CSV file into a DataFrame.
df = pd.read_csv('sample_data.csv')
print("\nLoaded DataFrame:")
print(df)
print("\n" + "="*50 + "\n")

# 4. Display the first few records to understand the structure.
print("First 3 records of the DataFrame:")
print(df.head(3))
print("\n" + "="*50 + "\n")

# 5. Get a summary of the dataset.
print("Summary of the DataFrame:")
df.info()
print("\n" + "="*50 + "\n")

# Step 3: Perform NumPy Operations
# Objective: Utilize NumPy for basic numerical operations and array manipulations.

# 6. Convert a DataFrame column to a NumPy array and perform array operations like mean and sum.
col2_numpy = df['col2'].to_numpy()
print("col2 as NumPy array:", col2_numpy)
mean_col2 = np.nanmean(col2_numpy) # Use nanmean to handle potential NaNs
sum_col1 = np.sum(df['col1'].to_numpy())
print("Mean of col2:", mean_col2)
print("Sum of col1:", sum_col1)
print("\n" + "="*50 + "\n")

# 7. Create a NumPy array and calculate the variance and standard deviation.
numpy_array = np.array([10, 20, 30, 40, 50])
variance_array = np.var(numpy_array)
std_dev_array = np.std(numpy_array)
print("NumPy array:", numpy_array)
print("Variance:", variance_array)
print("Standard Deviation:", std_dev_array)
print("\n" + "="*50 + "\n")

# 8. Use NumPy to filter based on conditions.
filtered_array = numpy_array[numpy_array > 30]
print("Filtered NumPy array (elements > 30):", filtered_array)
print("\n" + "="*50 + "\n")

# Step 4: Data Manipulation with Pandas
# Objective: Use Pandas to clean and manipulate dataset for analysis.

# 9. Handle missing data by filling or dropping.
print("Original DataFrame:")
print(df)
print("\nMissing values count per column:")
print(df.isnull().sum())

# Fill NaN in 'col2' with the mean
mean_col2_df = df['col2'].mean()
df_filled = df.fillna({'col2': mean_col2_df})
print("\nDataFrame after filling NaN in 'col2':")
print(df_filled)

# Alternatively, drop rows with NaN
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with NaN:")
print(df_dropped)
print("\n" + "="*50 + "\n")

# Let's work with the filled DataFrame for the next steps
df = df_filled.copy()

# 10. Create new columns or modify existing ones.
df['col1_squared'] = df['col1'] ** 2
df['col1_plus_10'] = df['col1'] + 10
print("DataFrame with new columns:")
print(df)
print("\n" + "="*50 + "\n")

# 11. Use groupby to aggregate data.
grouped_data = df.groupby('col3')['col1'].mean()
print("Average of col1 grouped by col3:")
print(grouped_data)
print("\n" + "="*50 + "\n")

# Step 5: Data Visualization with Matplotlib & Seaborn
# Objective: Visualize the data to identify patterns, trends, and insights.

# 12. Use Matplotlib to create a basic plot.
plt.figure(figsize=(6, 4))
plt.plot(df['col1'], df['col2'], marker='o', linestyle='-', color='blue')
plt.title('Col1 vs Col2')
plt.xlabel('Col1')
plt.ylabel('Col2')
plt.grid(True)
plt.show()

# 13. Create a histogram using Seaborn.
plt.figure(figsize=(6, 4))
sns.histplot(df['col1'], bins=5, kde=True)
plt.title('Histogram of Col1')
plt.xlabel('Col1 Value')
plt.ylabel('Frequency')
plt.show()

# 14. Plot a box plot for a clear view of data distribution.
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['col2'])
plt.title('Box Plot of Col2')
plt.ylabel('Col2 Value')
plt.show()
```
