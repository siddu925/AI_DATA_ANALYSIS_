### Monitor Data Quality Trends Over Time

**Task 1**: Create a Trends Analysis Report

**Objective**: Understand long-term data quality trends.

**Steps**:
1. Use historical data (or simulate data) to analyze how data quality has changed over time.
2. Calculate trends for the KPIs defined earlier using statistical measures or visual charts.
3. Write a report summarizing your findings, noting any persistent issues or improvements.

```python
# Write your code from here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# 1. Simulate historical data
np.random.seed(42)  # for reproducibility
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)
num_days = (end_date - start_date).days + 1
dates = [start_date + timedelta(days=i) for i in range(num_days)]

# Simulate daily KPI values with some trends
valid_product_id_percentage = 90 + np.sin(np.linspace(0, 4 * np.pi, num_days)) * 5 + np.random.normal(0, 1, num_days)
complete_shipping_percentage = 85 + np.cos(np.linspace(0, 4 * np.pi, num_days)) * 3 + np.random.normal(0, 2, num_days)
average_time_lag = 2.5 + np.linspace(0, 0.5, num_days) + np.random.normal(0, 0.1, num_days)

# Ensure percentages stay within 0-100 range
valid_product_id_percentage = np.clip(valid_product_id_percentage, 80, 100)
complete_shipping_percentage = np.clip(complete_shipping_percentage, 75, 95)

# Create a Pandas DataFrame
trend_data = pd.DataFrame({
    'Date': dates,
    'Valid Product IDs (%)': valid_product_id_percentage,
    'Complete Shipping (%)': complete_shipping_percentage,
    'Average Time Lag (Days)': average_time_lag
})

# Set 'Date' as index for time series analysis
trend_data.set_index('Date', inplace=True)

# 2. Calculate and visualize trends

# a) Time series plots
plt.figure(figsize=(12, 6))
plt.plot(trend_data.index, trend_data['Valid Product IDs (%)'], label='Valid Product IDs', color='skyblue')
plt.plot(trend_data.index, trend_data['Complete Shipping (%)'], label='Complete Shipping', color='lightcoral')
plt.xlabel('Date')
plt.ylabel('Percentage (%)')
plt.title('Data Quality Trends Over Time (Percentages)')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(trend_data.index, trend_data['Average Time Lag (Days)'], label='Average Time Lag', color='lightgreen')
plt.xlabel('Date')
plt.ylabel('Days')
plt.title('Data Quality Trend Over Time (Average Time Lag)')
plt.legend()
plt.grid(True)
plt.show()

# b) Rolling averages to smooth out noise and see longer-term trends
rolling_window = 30  # 30-day rolling average

plt.figure(figsize=(12, 6))
plt.plot(trend_data.index, trend_data['Valid Product IDs (%)'].rolling(window=rolling_window).mean(), label=f'{rolling_window}-Day Rolling Avg. (Valid IDs)', color='blue')
plt.plot(trend_data.index, trend_data['Complete Shipping (%)'].rolling(window=rolling_window).mean(), label=f'{rolling_window}-Day Rolling Avg. (Shipping)', color='red')
plt.xlabel('Date')
plt.ylabel('Percentage (%)')
plt.title(f'{rolling_window}-Day Rolling Average of Data Quality Metrics')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(trend_data.index, trend_data['Average Time Lag (Days)'].rolling(window=rolling_window).mean(), label=f'{rolling_window}-Day Rolling Avg. (Time Lag)', color='green')
plt.xlabel('Date')
plt.ylabel('Days')
plt.title(f'{rolling_window}-Day Rolling Average of Average Time Lag')
plt.legend()
plt.grid(True)
plt.show()

# 3. Simulate a report (printing findings)
print("\n--- Data Quality Trend Analysis Report ---")
print("\nKey Observations:")

# Analyze trends (simple examples)
if trend_data['Valid Product IDs (%)'].iloc[-1] < trend_data['Valid Product IDs (%)'].iloc[0]:
    print("- The percentage of orders with valid product IDs shows a slight decline over the year.")
else:
    print("- The percentage of orders with valid product IDs has remained relatively stable or slightly improved over the year.")

if trend_data['Complete Shipping (%)'].iloc[-1] < trend_data['Complete Shipping (%)'].iloc[0]:
    print("- The percentage of orders with complete shipping information indicates a minor downward trend.")
else:
    print("- The percentage of orders with complete shipping information has been generally stable or slightly increasing.")

if trend_data['Average Time Lag (Days)'].iloc[-1] > trend_data['Average Time Lag (Days)'].iloc[0]:
    print("- The average time lag between order placement and shipment has increased over the year, which might require investigation.")
else:
    print("- The average time lag between order placement and shipment has remained stable or decreased.")

# Identify potential persistent issues or improvements (based on simulated trends)
print("\nPotential Persistent Issues or Improvements:")
if trend_data['Average Time Lag (Days)'].mean() > 3:
    print("- The average shipment time consistently remains above 3 days, suggesting a potential bottleneck in the fulfillment process.")

if trend_data['Valid Product IDs (%)'].min() < 85:
    print("- There were periods where the percentage of valid product IDs dropped below 85%, indicating potential data integrity issues that need attention.")

print("\nRecommendations:")
print("- Further investigation is recommended into the increasing average time lag.")
print("- Monitoring the percentage of valid product IDs is crucial to ensure data accuracy.")
```
